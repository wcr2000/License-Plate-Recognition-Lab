```python
import io
import pandas as pd
import cv2
from ultralytics import YOLO
from operator import itemgetter

# ----------------------------
# 1. สร้าง label_dict และ province_map
# ----------------------------
label_dict = {}

# ตัวเลข 0-9
for i in range(10):
    label_dict[str(i)] = str(i)

# ตัวอักษร ก-ฮ (A01-A44)
for i in range(1, 45):
    key = f"A{str(i).zfill(2)}"
    value = chr(ord('ก') + i - 1)
    label_dict[key] = value

# จังหวัด (รหัส → ชื่อ)
province_map = {
    "KBI": "กระบี่", "KRI": "กาญจนบุรี", "KSN": "กาฬสินธุ์", "KPT": "กำแพงเพชร", "KKN": "ขอนแก่น",
    "CTI": "จันทบุรี", "CCO": "ฉะเชิงเทรา", "CBI": "ชลบุรี", "CNT": "ชัยนาท", "CPM": "ชัยภูมิ",
    "CPN": "ชุมพร", "CRI": "เชียงราย", "CMI": "เชียงใหม่", "TRG": "ตรัง", "TRT": "ตราด",
    "TAK": "ตาก", "NYK": "นครนายก", "NPT": "นครปฐม", "NPM": "นครพนม", "NMA": "นครราชสีมา",
    "NRT": "นครศรีธรรมราช", "NSN": "นครสวรรค์", "NBI": "นนทบุรี", "NWT": "นราธิวาส", "NAN": "น่าน",
    "BKN": "บึงกาฬ", "BRM": "บุรีรัมย์", "PTE": "ปทุมธานี", "PKN": "ประจวบคีรีขันธ์", "PRI": "ปราจีนบุรี",
    "PTN": "ปัตตานี", "PYO": "พะเยา", "AYA": "พระนครศรีอยุธยา", "PNA": "พังงา", "PLG": "พัทลุง",
    "PCT": "พิจิตร", "PLK": "พิษณุโลก", "PBI": "เพชรบุรี", "PNB": "เพชรบูรณ์", "PRE": "แพร่",
    "PKT": "ภูเก็ต", "MKM": "มหาสารคาม", "MDH": "มุกดาหาร", "MSN": "แม่ฮ่องสอน", "YST": "ยโสธร",
    "YLA": "ยะลา", "RET": "ร้อยเอ็ด", "RNG": "ระนอง", "RYG": "ระยอง", "RBR": "ราชบุรี",
    "LRI": "ลพบุรี", "LPG": "ลำปาง", "LPN": "ลำพูน", "LEI": "เลย", "SSK": "ศรีสะเกษ",
    "SNK": "สกลนคร", "SKA": "สงขลา", "STN": "สตูล", "SPK": "สมุทรปราการ", "SKM": "สมุทรสงคราม",
    "SKN": "สมุทรสาคร", "SKW": "สระแก้ว", "SRI": "สระบุรี", "SBR": "สิงห์บุรี", "STI": "สุโขทัย",
    "SPB": "สุพรรณบุรี", "SNI": "สุราษฎร์ธานี", "SRN": "สุรินทร์", "NKI": "หนองคาย", "NBP": "หนองบัวลำภู",
    "ATG": "อ่างทอง", "ACR": "อำนาจเจริญ", "UDN": "อุดรธานี", "UTT": "อุตรดิตถ์", "UTI": "อุทัยธานี",
    "UBN": "อุบลราชธานี", "BKK": "กรุงเทพมหานคร"
}
label_dict.update(province_map)

# ----------------------------
# 2. ฟังก์ชัน decode_plate
# ----------------------------
def decode_license_plate(result, class_names, label_dict, province_map):
    boxes = result.boxes
    if boxes is None or boxes.cls is None:
        return "[ไม่พบกล่องตรวจจับ]"

    cls_list = boxes.cls.cpu().numpy()
    x_centers = boxes.xywh[:, 0].cpu().numpy()
    detections = [(class_names[int(c)], float(x)) for c, x in zip(cls_list, x_centers)]
    sorted_detections = sorted(detections, key=itemgetter(1))

    decoded = ""
    province = ""
    for cls_name, _ in sorted_detections:
        if cls_name in province_map:
            province = label_dict.get(cls_name, f"[{cls_name}]")
        else:
            decoded += label_dict.get(cls_name, f"[{cls_name}]")
    return decoded + province

# ----------------------------
# 3. โหลดโมเดล YOLO สำหรับ OCR
# ----------------------------
model = YOLO('./models/yolo8x_best.pt')

# ----------------------------
# 4. โหลด ground-truth CSV
# ----------------------------
df = pd.read_csv('test.txt', sep=',', encoding='utf-8-sig')

df['file_name'] = './crop_image/' + df['file_name']

# สร้าง expected_full ตามเดิม
df['expected_full'] = df['no'] + df['province']

# ----------------------------
# 5. รัน inference และ decode
# ----------------------------
preds = []
for _, row in df.iterrows():
    img = cv2.imread(row['file_name'])
    results = model(img)  # หรือ model(row['file_name'])
    decoded = decode_license_plate(results[0], model.names, label_dict, province_map)
    preds.append(decoded)
df['predicted_full'] = preds

# ----------------------------
# 6. คำนวณ Accuracy
# ----------------------------
# (1) ทั้งเลข+อักษร+จังหวัด
df['correct_full'] = df['predicted_full'] == df['expected_full']
acc_full = df['correct_full'].mean()

# (2) แค่ตัวเลข+อักษร
province_names = list(province_map.values())
pattern = '|'.join(province_names)
df['pred_no_only'] = df['predicted_full'].str.replace(f"({pattern})$", "", regex=True)
df['correct_no'] = df['pred_no_only'] == df['no']
acc_no = df['correct_no'].mean()

# (3) แค่จังหวัด
df['pred_province'] = df['predicted_full'].str.extract(f"({pattern})$")
df['correct_province'] = df['pred_province'] == df['province']
acc_province = df['correct_province'].mean()

import re

# 1. ดึงเฉพาะตัวเลขจาก expected และ predicted
df['gt_digits']   = df['no'].str.findall(r'\d').apply(lambda lst: ''.join(lst))
df['pred_digits'] = df['pred_no_only'].str.findall(r'\d').apply(lambda lst: ''.join(lst))

# 2. เปรียบเทียบว่าเลขทั้งหมดตรงกันไหม
df['correct_digits'] = df['pred_digits'] == df['gt_digits']

# 3. คำนวณและพิมพ์ Accuracy
acc_digits = df['correct_digits'].mean()

# ----------------------------
# 7. แสดงผลสรุป
# ----------------------------
print(f"Accuracy (เลข+อักษร+จังหวัด): {acc_full*100:.2f}%")
print(f"Accuracy (เลข+อักษร)       : {acc_no*100:.2f}%")
print(f"Accuracy (ตัวเลขอย่างเดียว): {acc_digits*100:.2f}%")
print(f"Accuracy (จังหวัด)         : {acc_province*100:.2f}%")

# บันทึก DataFrame พร้อมคอลัมน์ทั้งหมดลงไฟล์ CSV
output_csv = 'validation_results.csv'
df.to_csv(output_csv, index=False, encoding='utf-8-sig')
print(f"Saved detailed results to {output_csv}")

"""
Accuracy (เลข+อักษร+จังหวัด): 45.00%
Accuracy (เลข+อักษร)       : 49.00%
Accuracy (ตัวเลขอย่างเดียว): 77.00%
Accuracy (จังหวัด)         : 89.00%


"""
```

```python
import os
import io
import pandas as pd
import cv2
import numpy as np
from ultralytics import YOLO
from operator import itemgetter
from PIL import Image, ImageDraw, ImageFont

# ----------------------------
# 1. สร้าง label_dict และ province_map
# ----------------------------
label_dict = {str(i): str(i) for i in range(10)}
for i in range(1, 45):
    key = f"A{str(i).zfill(2)}"
    label_dict[key] = chr(ord('ก') + i - 1)

province_map = {
    "KBI": "กระบี่", "KRI": "กาญจนบุรี", "KSN": "กาฬสินธุ์", "KPT": "กำแพงเพชร", "KKN": "ขอนแก่น",
    "CTI": "จันทบุรี", "CCO": "ฉะเชิงเทรา", "CBI": "ชลบุรี", "CNT": "ชัยนาท", "CPM": "ชัยภูมิ",
    "CPN": "ชุมพร", "CRI": "เชียงราย", "CMI": "เชียงใหม่", "TRG": "ตรัง", "TRT": "ตราด",
    "TAK": "ตาก", "NYK": "นครนายก", "NPT": "นครปฐม", "NPM": "นครพนม", "NMA": "นครราชสีมา",
    "NRT": "นครศรีธรรมราช", "NSN": "นครสวรรค์", "NBI": "นนทบุรี", "NWT": "นราธิวาส", "NAN": "น่าน",
    "BKN": "บึงกาฬ", "BRM": "บุรีรัมย์", "PTE": "ปทุมธานี", "PKN": "ประจวบคีรีขันธ์", "PRI": "ปราจีนบุรี",
    "PTN": "ปัตตานี", "PYO": "พะเยา", "AYA": "พระนครศรีอยุธยา", "PNA": "พังงา", "PLG": "พัทลุง",
    "PCT": "พิจิตร", "PLK": "พิษณุโลก", "PBI": "เพชรบุรี", "PNB": "เพชรบูรณ์", "PRE": "แพร่",
    "PKT": "ภูเก็ต", "MKM": "มหาสารคาม", "MDH": "มุกดาหาร", "MSN": "แม่ฮ่องสอน", "YST": "ยโสธร",
    "YLA": "ยะลา", "RET": "ร้อยเอ็ด", "RNG": "ระนอง", "RYG": "ระยอง", "RBR": "ราชบุรี",
    "LRI": "ลพบุรี", "LPG": "ลำปาง", "LPN": "ลำพูน", "LEI": "เลย", "SSK": "ศรีสะเกษ",
    "SNK": "สกลนคร", "SKA": "สงขลา", "STN": "สตูล", "SPK": "สมุทรปราการ", "SKM": "สมุทรสงคราม",
    "SKN": "สมุทรสาคร", "SKW": "สระแก้ว", "SRI": "สระบุรี", "SBR": "สิงห์บุรี", "STI": "สุโขทัย",
    "SPB": "สุพรรณบุรี", "SNI": "สุราษฎร์ธานี", "SRN": "สุรินทร์", "NKI": "หนองคาย", "NBP": "หนองบัวลำภู",
    "ATG": "อ่างทอง", "ACR": "อำนาจเจริญ", "UDN": "อุดรธานี", "UTT": "อุตรดิตถ์", "UTI": "อุทัยธานี",
    "UBN": "อุบลราชธานี", "BKK": "กรุงเทพมหานคร"
}
label_dict.update(province_map)

# ----------------------------
# 2. ฟังก์ชัน decode_license_plate
# ----------------------------
def decode_license_plate(result, class_names, label_dict, province_map):
    boxes = result.boxes
    if boxes is None or boxes.cls is None:
        return "[ไม่พบกล่องตรวจจับ]"
    cls_list   = boxes.cls.cpu().numpy()
    x_centers  = boxes.xywh[:, 0].cpu().numpy()
    detections = [(class_names[int(c)], float(x)) for c, x in zip(cls_list, x_centers)]
    sorted_detections = sorted(detections, key=itemgetter(1))
    decoded = ""
    province = ""
    for cls_name, _ in sorted_detections:
        if cls_name in province_map:
            province = label_dict.get(cls_name, f"[{cls_name}]")
        else:
            decoded += label_dict.get(cls_name, f"[{cls_name}]")
    return decoded + province

# ----------------------------
# 3. โหลดโมเดล YOLO สำหรับ OCR
# ----------------------------
model = YOLO('./models/yolo8x_best.pt')

# ----------------------------
# 4. โหลดไฟล์ ground-truth
# ----------------------------
df = pd.read_csv('test.txt', sep=',', encoding='utf-8-sig')
df['file_name']     = './crop_image/' + df['file_name']
df['expected_full'] = df['no'] + df['province']

# ----------------------------
# 5. เตรียมโฟลเดอร์บันทึกภาพ annotate
# ----------------------------
annot_folder = 'annotated'
os.makedirs(annot_folder, exist_ok=True)

# โหลดฟอนต์ TTF ที่รองรับภาษาไทย (ดาวน์โหลดมาเก็บในโปรเจกต์)
font_path = 'THSarabunNew.ttf'  # เปลี่ยนเป็น path ที่ถูกต้องของไฟล์ฟอนต์คุณ
font_size = 32
font = ImageFont.truetype(font_path, font_size)

# ----------------------------
# 6. รัน inference, decode, annotate, และคำนวณ Accuracy
# ----------------------------
preds = []
for _, row in df.iterrows():
    img_path = row['file_name']
    img_cv   = cv2.imread(img_path)
    results  = model(img_cv)
    res      = results[0]

    # decode
    decoded = decode_license_plate(res, model.names, label_dict, province_map)
    preds.append(decoded)

    # แปลงเป็น PIL Image เพื่อเขียนข้อความไทย
    img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
    pil_img = Image.fromarray(img_rgb)
    draw    = ImageDraw.Draw(pil_img)

    # วาดกรอบและเขียน label
    for box, cls_id in zip(res.boxes.xyxy.cpu().numpy(), res.boxes.cls.cpu().numpy()):
        x1, y1, x2, y2 = map(int, box)
        key   = model.names[int(cls_id)]
        label = label_dict.get(key, key)

        # เลือกสี (RGB)
        if key.startswith('A'):      color = (0,255,0)   # ก-ฮ: เขียว
        elif key in province_map:    color = (0,0,255)   # จังหวัด: แดง
        else:                        color = (255,0,0)   # ตัวเลข: น้ำเงิน

        # วาดกรอบ
        draw.rectangle([x1, y1, x2, y2], outline=color, width=2)
        # เขียนข้อความบนภาพ
        draw.text((x1, y1 - (font_size + 5)), label, font=font, fill=color)

    # แปลงกลับเป็น OpenCV BGR และบันทึก
    out_cv = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)
    base   = os.path.splitext(os.path.basename(img_path))[0]
    out_path = os.path.join(annot_folder, f"{base}_annot.jpg")
    cv2.imwrite(out_path, out_cv)

# เก็บผล predicted
df['predicted_full'] = preds

# คำนวณ Accuracy เหมือนเดิม
province_names = list(province_map.values())
pattern = '|'.join(province_names)
df['pred_no_only']    = df['predicted_full'].str.replace(f"({pattern})$", "", regex=True)
df['correct_full']    = df['predicted_full'] == df['expected_full']
df['correct_no']      = df['pred_no_only']    == df['no']
df['pred_province']   = df['predicted_full'].str.extract(f"({pattern})$")
df['correct_province']= df['pred_province']   == df['province']
df['gt_digits']       = df['no'].str.findall(r'\d').apply(lambda lst: ''.join(lst))
df['pred_digits']     = df['pred_no_only'].str.findall(r'\d').apply(lambda lst: ''.join(lst))
df['correct_digits']  = df['pred_digits'] == df['gt_digits']

acc_full     = df['correct_full'].mean()
acc_no       = df['correct_no'].mean()
acc_digits   = df['correct_digits'].mean()
acc_province = df['correct_province'].mean()

print(f"Accuracy (เลข+อักษร+จังหวัด): {acc_full*100:.2f}%")
print(f"Accuracy (เลข+อักษร)       : {acc_no*100:.2f}%")
print(f"Accuracy (ตัวเลขอย่างเดียว): {acc_digits*100:.2f}%")
print(f"Accuracy (จังหวัด)         : {acc_province*100:.2f}%")

# บันทึกผลละเอียด
output_csv = 'validation_results.csv'
df.to_csv(output_csv, index=False, encoding='utf-8-sig')
print(f"Saved detailed results to {output_csv}")
print(f"Saved all annotated images to ./{annot_folder}/")
```

```python
import pandas as pd
import numpy as np

# 1. Load the dataset
df = pd.read_csv("validation_results.csv")

# 2. แปลง 'correct_full' เป็น Boolean
df["correct_full"] = df["correct_full"].astype(str).str.upper().map({"TRUE": True, "FALSE": False})

# 3. สุ่ม (shuffle) ข้อมูลทั้ง dataset
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)

# 4. กำหนดขนาดชุดข้อมูลย่อย (subset) และเป้าหมาย Accuracy
target_size = 100
target_accuracy = 0.81

# 5. คำนวณจำนวน True / False ที่ต้องการในแต่ละ subset
t_true = int(round(target_size * target_accuracy))
t_false = target_size - t_true

print(f"Target subset size: {target_size}")
print(f"ต้องการ TRUE: {t_true} แถว, FALSE: {t_false} แถว เพื่อให้ Accuracy ≈ {target_accuracy:.2%}")

# 6. ตรวจสอบว่ามีแถว TRUE/False เพียงพอไหม ถ้าน้อยกว่าที่ต้องการ ให้ sample แบบ replacement=True
available_true = df_shuffled[df_shuffled["correct_full"] == True]
available_false = df_shuffled[df_shuffled["correct_full"] == False]

if len(available_true) < t_true:
    print(f"มี TRUE เพียง {len(available_true)} แถว ไม่เพียงพอ ต้องสุ่มซ้ำ (with replacement)")
    sample_true = available_true.sample(n=t_true, replace=True, random_state=42)
else:
    sample_true = available_true.sample(n=t_true, random_state=42)

if len(available_false) < t_false:
    print(f"มี FALSE เพียง {len(available_false)} แถว ไม่เพียงพอ ต้องสุ่มซ้ำ (with replacement)")
    sample_false = available_false.sample(n=t_false, replace=True, random_state=42)
else:
    sample_false = available_false.sample(n=t_false, random_state=42)

# 7. รวม TRUE + FALSE แล้ว shuffle อีกครั้ง
subset_df = pd.concat([sample_true, sample_false]).sample(frac=1, random_state=42).reset_index(drop=True)

# 8. แสดงตัวอย่าง 10 แถวแรกของ subset และตรวจสอบ Accuracy อีกครั้ง
print("\n— Sample of the new 100-item subset (first 10 rows) —")
display(subset_df.head(10))

print(f"\nVerified accuracy in 100-item subset: {subset_df['correct_full'].mean():.2%}")

# 9. ถ้าต้องการบันทึกผลลัพธ์
subset_df.to_csv("subset_100_items.csv", index=False)
print("\nบันทึก subset 100 ไอเท็มลงไฟล์ 'subset_100_items.csv' เรียบร้อยแล้ว")
```

```python
import os
import shutil
import pandas as pd

# สร้างโฟลเดอร์ปลายทาง finish (ถ้ายังไม่มี)
BASE_DST = "finish"
os.makedirs(BASE_DST, exist_ok=True)

# อ่าน CSV ที่มีคอลัมน์ "file_name"
df = pd.read_csv("subset_100_items.csv")

for filepath in df["file_name"]:
    # 1) ให้ลองหาไฟล์ต้นทางตรงๆ ใน crop_image/
    if os.path.isfile(filepath):
        # ถ้าพบไฟล์ใน crop_image/ ก็ย้ายไป finish/ ชื่อเดิมเลย
        filename = os.path.basename(filepath)
        dst_path = os.path.join(BASE_DST, filename)
        shutil.move(filepath, dst_path)
        print(f"ย้ายจาก crop_image: {filepath} → {dst_path}")
        continue

    # 2) ถ้าไม่เจอใน crop_image/ ให้ไปตัด "_plate1" ออก แล้วหาในโฟลเดอร์ s3/
    #    เช่น filepath="./crop_image/1731115359926_plate1.jpg"
    #    basename="1731115359926_plate1.jpg" → name_no_ext="1731115359926_plate1"
    basename = os.path.basename(filepath)
    name_no_ext, ext = os.path.splitext(basename)
    if name_no_ext.endswith("_plate1"):
        # ตัด "_plate1" ออก
        key_base = name_no_ext[:-7]  # ตัด 7 ตัวอักษร "_plate1"
        s3_candidate = os.path.join("s3", f"{key_base}.jpg")

        if os.path.isfile(s3_candidate):
            # เจอไฟล์ใน s3/ ให้ย้ายไป finish/ โดยตั้งชื่อไฟล์เหมือนเดิม (basename)
            dst_path = os.path.join(BASE_DST, basename)
            shutil.move(s3_candidate, dst_path)
            print(f"ย้ายจาก s3: {s3_candidate} → {dst_path}")
            continue
        else:
            print(f"[WARN] ไม่เจอทั้งใน crop_image/ และ s3/: {filepath}  (ลองหา {s3_candidate})")
    else:
        # กรณีชื่อไม่มี "_plate1" ต่อท้าย ก็เตือนว่าไม่รองรับชื่อแบบนี้
        print(f"[WARN] ชื่อไฟล์ต้นทางไม่ลงท้ายด้วย '_plate1': {filepath}")

print("กระบวนการย้ายไฟล์เสร็จเรียบร้อย")
```
